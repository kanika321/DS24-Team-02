```python
import os

def get_all_parquet_paths(root_dir):
    parquet_files_100KHz = []
    parquet_files_2000KHz = []
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith('.parquet'):
                if 'Sampling100KHz' in file:
                    parquet_files_100KHz.append(os.path.join(root, file))
                elif 'Sampling2000KHz' in file:
                    parquet_files_2000KHz.append(os.path.join(root, file))
    return parquet_files_100KHz, parquet_files_2000KHz

# Example usage:
nok_root_dir = r"C:\Users\Vaishnavi Mall\Desktop\Sem3\DataScience\Data\NOK_Measurements"
ok_root_dir = r"C:\Users\Vaishnavi Mall\Desktop\Sem3\DataScience\Data\OK_Measurements"

nok_100KHz_files, nok_2000KHz_files = get_all_parquet_paths(nok_root_dir)
ok_100KHz_files, ok_2000KHz_files = get_all_parquet_paths(ok_root_dir)

print("NOK 100KHz files:", nok_100KHz_files)
print("NOK 2000KHz files:", nok_2000KHz_files)
print("OK 100KHz files:", ok_100KHz_files)
print("OK 2000KHz files:", ok_2000KHz_files)
```
