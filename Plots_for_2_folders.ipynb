```python
import pandas as pd

# Load data from the start and end folders
start_folder_1 = '3.csv'
start_folder_2 = '4.csv'
end_folder_1 = '1.csv'
end_folder_2 = '2.csv'

# Read the CSV files
start_data_1 = pd.read_csv(start_folder_1)
start_data_2 = pd.read_csv(start_folder_2)
end_data_1 = pd.read_csv(end_folder_1)
end_data_2 = pd.read_csv(end_folder_2)

# Combine the start and end data based on the file names (assuming same structure and length)
combined_data_1 = pd.concat([start_data_1, end_data_1], keys=['start', 'end'])
combined_data_2 = pd.concat([start_data_2, end_data_2], keys=['start', 'end'])

# Calculate differences between start and end for each parameter
diff_data_1 = end_data_1 - start_data_1
diff_data_2 = end_data_2 - start_data_2

# For a more detailed analysis, you might look at statistics or plot the differences
import matplotlib.pyplot as plt

# Plot differences for the parameters in 1.csv
for column in diff_data_1.columns:
    plt.figure()
    plt.plot(diff_data_1[column])
    plt.xlabel('Sample Index')
    plt.ylabel(f'Difference in {column}')
    plt.title(f'Difference in {column} over 30 seconds')
    plt.show()

# Plot differences for the parameters in 2.csv
plt.figure()
plt.plot(diff_data_2['AEKi_rate2000000_clipping0_batch0'])
plt.xlabel('Sample Index')
plt.ylabel('Difference in AEKi')
plt.title('Difference in AEKi over 30 seconds')
plt.show()
```
