Anomaly detection


```python
# Define window size for rolling statistics
window_size = 1000

# Calculate rolling mean and standard deviation
df['Irms_Grinding_rolling_mean'] = df['Irms_Grinding_rate100000_clipping0_batch0'].rolling(window=window_size).mean()
df['Irms_Grinding_rolling_std'] = df['Irms_Grinding_rate100000_clipping0_batch0'].rolling(window=window_size).std()

df['Spindle_Current_L1_rolling_mean'] = df['Grinding spindle current L1_rate100000_clipping0_batch0'].rolling(window=window_size).mean()
df['Spindle_Current_L1_rolling_std'] = df['Grinding spindle current L1_rate100000_clipping0_batch0'].rolling(window=window_size).std()

df['Spindle_Current_L2_rolling_mean'] = df['Grinding spindle current L2_rate100000_clipping0_batch0'].rolling(window=window_size).mean()
df['Spindle_Current_L2_rolling_std'] = df['Grinding spindle current L2_rate100000_clipping0_batch0'].rolling(window=window_size).std()

df['Spindle_Current_L3_rolling_mean'] = df['Grinding spindle current L3_rate100000_clipping0_batch0'].rolling(window=window_size).mean()
df['Spindle_Current_L3_rolling_std'] = df['Grinding spindle current L3_rate100000_clipping0_batch0'].rolling(window=window_size).std()
```

```python
# Define a threshold for anomaly detection (e.g., 3 standard deviations)
threshold = 3/1

# Identify anomalies
df['Irms_Grinding_anomaly'] = abs(df['Irms_Grinding_rate100000_clipping0_batch0'] - df['Irms_Grinding_rolling_mean']) > (threshold * df['Irms_Grinding_rolling_std'])
df['Spindle_Current_L1_anomaly'] = abs(df['Grinding spindle current L1_rate100000_clipping0_batch0'] - df['Spindle_Current_L1_rolling_mean']) > (threshold * df['Spindle_Current_L1_rolling_std'])
df['Spindle_Current_L2_anomaly'] = abs(df['Grinding spindle current L2_rate100000_clipping0_batch0'] - df['Spindle_Current_L2_rolling_mean']) > (threshold * df['Spindle_Current_L2_rolling_std'])
df['Spindle_Current_L3_anomaly'] = abs(df['Grinding spindle current L3_rate100000_clipping0_batch0'] - df['Spindle_Current_L3_rolling_mean']) > (threshold * df['Spindle_Current_L3_rolling_std'])
```

```python
# Plotting function
def plot_with_anomalies(df, column, anomaly_column, title, ax):
    sns.lineplot(data=df, x=df.index, y=column, ax=ax, label='Current')
    sns.scatterplot(data=df[df[anomaly_column]], x=df[df[anomaly_column]].index, y=column, ax=ax, color='red', label='Anomaly', s=10)
    ax.set_title(title)
    ax.set_ylabel('Current (A)')
    ax.legend()

# Set the size of the plot
fig, axs = plt.subplots(4, 1, figsize=(14, 16), sharex=True)

# Plot the RMS grinding current with anomalies
plot_with_anomalies(df, 'Irms_Grinding_rate100000_clipping0_batch0', 'Irms_Grinding_anomaly', 'Irms Grinding with Anomalies', axs[0])

# Plot Grinding spindle current L1 with anomalies
plot_with_anomalies(df, 'Grinding spindle current L1_rate100000_clipping0_batch0', 'Spindle_Current_L1_anomaly', 'Grinding Spindle Current L1 with Anomalies', axs[1])

# Plot Grinding spindle current L2 with anomalies
plot_with_anomalies(df, 'Grinding spindle current L2_rate100000_clipping0_batch0', 'Spindle_Current_L2_anomaly', 'Grinding Spindle Current L2 with Anomalies', axs[2])

# Plot Grinding spindle current L3 with anomalies
plot_with_anomalies(df, 'Grinding spindle current L3_rate100000_clipping0_batch0', 'Spindle_Current_L3_anomaly', 'Grinding Spindle Current L3 with Anomalies', axs[3])

# Set labels
plt.xlabel('Sample Index')
plt.tight_layout()
plt.show()
```
